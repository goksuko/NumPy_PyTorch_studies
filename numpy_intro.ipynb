```python
# additional code from example

import numpy as np
import matplotlib.pyplot as plt

# Function to generate Fibonacci sequence
def fibonacci(n):
    fib_sequence = [0, 1]
    for i in range(2, n):
        fib_sequence.append(fib_sequence[i-1] + fib_sequence[i-2])
    return fib_sequence

# Generate first 100 Fibonacci numbers
n = 100
fib_numbers = fibonacci(n)

# Create plot
plt.figure(figsize=(12, 6))
plt.plot(range(n), fib_numbers, marker='o', linestyle='-', markersize=4)
plt.title('First 100 Fibonacci Numbers')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)

# Add a log scale y-axis since Fibonacci numbers grow exponentially
plt.yscale('log')
plt.tight_layout()

# Show the plot
plt.show()

# Print the first few and last few numbers for reference
print("First 10 Fibonacci numbers:", fib_numbers[:10])
print("Last 5 Fibonacci numbers:", fib_numbers[-5:])
```
